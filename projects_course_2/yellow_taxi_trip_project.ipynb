# NYC Yellow Taxi Trip Data Cleaning and Exploratory Analysis (Python)

```python
# Import necessary libraries
import pandas as pd
import numpy as np
```

```python
# Load the dataset and display the first 10 rows
df = pd.read_csv('taxi_data.csv')
df.head(10)   # By default, displays first 5 rows
```

```python
# Concise summary of the DataFrame
df.info()
```

```python
# summary statistics of the numerical columns
df.describe()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# Drop Unnamed column as it is not needed
df.drop(columns=['Unnamed: 0'], inplace=True)

# Check the first 5 rows again to confirm the column has been dropped
df.head()
```

```python
# Convert two 'object' type variables 'tpep_pickup_datetime', and 'tpep_dropoff_datetime' to datetime format
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

# Verify the conversion by checking the data types again
df.dtypes
```

```python
# Sort the variable 'trip_distance' in descending order and display the top 5 rows
# df.sort_values(by='trip_distance', ascending=False).head()
df_sorted = df.sort_values(by='trip_distance', ascending=False)
df_sorted.head(10)
```

```python
# Sort the second variable 'total_amount' in descending order and display the top 5 rows
# df.sort_values(by='total_amount', ascending=False).head()
df_sorted_amount = df.sort_values(['total_amount'], ascending=False)['total_amount']
df_sorted_amount.head(20)
```

```python
# Check the last 20 values of the 'total_amount' column to see if there are any negative or zero values
df_sorted_amount.tail(20)
```

```python
# How many of each payment type are there?
df['payment_type'].value_counts()
```

```python
# What is the average tip for trips paid by 'credit card'
average_cc_tip = df[df['payment_type'] == 1]['tip_amount'].mean()
print('Average Credit Card Tip:' , average_cc_tip)

# What is the average tip for trips paid by 'cash'
average_cash_tip = df[df['payment_type']==2]['tip_amount'].mean()
print('Average Cash Tip:' , average_cash_tip)
```

```python
# How many time of each VendorID represented in the data
df['VendorID'].value_counts()
```

```python
# Mean value of 'total_amount' for each vendor
df.groupby('VendorID').mean(numeric_only=True)[['total_amount']]
```

```python
# Filter the data for credit card payments only
credit_card = df[df['payment_type'] == 1]

# Filter the only credit card data with passengers count
credit_card['passenger_count'].value_counts()
```

```python
# Calculate the average tip amount for each passenger count in credit card payments
credit_card.groupby(['passenger_count']).mean(numeric_only=True)[['tip_amount']]
```
